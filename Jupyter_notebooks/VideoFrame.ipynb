```python
import glob
import os
import cv2
```

```python
image_folder = '.././Figures/60kV-0126/Movie_Prediction'
video_name = 'video.avi'

images = [img for img in sorted(glob.glob(image_folder +'/*.png'), key=numericalSort)] #os.listdir(image_folder) if img.endswith(".png")]
frame = cv2.imread(images[0])
height, width, layers = frame.shape
fourcc = cv2.VideoWriter_fourcc(*'3IVD')
video = cv2.VideoWriter(video_name, fourcc, 10, (width,height))
#video = cv2.VideoWriter(video_name, -1, 1, (width,height))

for i in range(len(images)):
    video.write(cv2.imread(images[i]))

cv2.destroyAllWindows()
video.release()
```

```python
import re
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```
